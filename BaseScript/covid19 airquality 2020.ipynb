```python
df=pd.read_csv("waqi-covid19-airqualitydata-2020.csv")
df.head()
```

```python
milan_df=df.loc[df['City']=="Milan",:]
milan_df.head()
```

```python
milanwoweather = milan_df.loc[(milan_df['Specie'] == "co") | (milan_df["Specie"] == "pm10") | 
                              (milan_df['Specie'] == "no2") | (milan_df['Specie'] == "o3") |
                              (milan_df['Specie'] == "pm25") | (milan_df['Specie'] == "so2"), :]

finalmilan = milanwoweather[["Date","Specie", "median"]].sort_values(by=['Date','Specie'])
finalmilan.head(20)
```

```python
grouped_finalmilan=finalmilan.groupby(["Date","Specie"])
grouped_finalmilan['median'].mean()
```
