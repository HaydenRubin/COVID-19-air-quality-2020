```python
wuhan19_csv = "CleanDatasets/wuhan2019.csv"
```

```python
wuhan2019 = pd.read_csv(wuhan19_csv)
wuhan2019.head()
```

```python
wuhan2019[['year','month','day']] = wuhan2019['date'].str.split('-',expand=True)
wuhan2019.head()
```

```python
wuhan2019['day'] = wuhan2019['day'].astype(str).astype('int64')
wuhan2019['year'] = wuhan2019['year'].astype(str).astype('int64')
wuhan2019['month'] = wuhan2019['month'].astype(str).astype(int)
```

```python
wuhan2019['AQI'].max()
```

```python
wuhan2019aq = wuhan2019.pivot_table('AQI','month','day')
```

```python
f, ax = plt.subplots(figsize=(13, 7))
sns.heatmap(wuhan2019aq, annot=False, fmt="d", linewidths=.5, ax=ax, vmax=300, cmap = "RdYlGn_r")
```

```python
wuhan_df = pd.read_csv("../CleanDatasets/wuhan_AQI.csv")
wuhan_df.head()
```

```python
wuhan_2019 =wuhan_df[(wuhan_df['date']>='2019-01-01') & (wuhan_df['date']<='2019-12-31')]
wuhan_2019 = wuhan_2019.sort_values('date')
wuhan_2019.head()
```

```python
wuhan_2019[['year','month','day']] = wuhan_2019['date'].str.split('-',expand=True)
wuhan_2019.head()
```

```python
wuhan_2019['day'] = wuhan_2019['day'].astype(str).astype('int64')
wuhan_2019['year'] = wuhan_2019['year'].astype(str).astype('int64')
wuhan_2019['month'] = wuhan_2019['month'].astype(str).astype(int)
```

```python
wuhan_2019['AQI'].max()
```

```python
wuhan_2019aq = wuhan_2019.pivot_table('AQI','month','day')
```

```python
f1, ax = plt.subplots(figsize=(20, 7))
sns.heatmap(wuhan_2019aq, annot=True, fmt="g", linewidths=.5, ax=ax, vmax=200, cmap = "RdYlGn_r")
plt.title("Wuhan, China Air Quality Index (AQI) Heatmap per day in 2019", fontsize=20)
plt.ylabel("Month", fontsize=15)
plt.xlabel("Day", fontsize=15)
#plt.savefig("2019_WuhanAQI_CalendarHeatmap.png")
```
